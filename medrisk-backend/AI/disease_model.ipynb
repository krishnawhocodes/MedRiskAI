```python
import pandas as pd
import joblib
import json
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

print("--- Starting Model Training ---")
```

```python
try:
    df = pd.read_csv("synthetic_blood_data.csv")
except FileNotFoundError:
    print("Error: synthetic_blood_data.csv not found!")
    exit()
```

```python
features = ["Hemoglobin", "WBC", "TSH", "MPV"]
target = "disease"

X = df[features]
y = df[target]
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = list(le.classes_)
```

```python
model_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('classifier', DecisionTreeClassifier(random_state=42))
])
```

```python
print(f"Training model on {len(X)} samples...")
model_pipeline.fit(X, y_encoded)
print("Model training complete.")
```

```python
model_filename = "disease_model.pkl"
joblib.dump(model_pipeline, model_filename)
print(f"Model saved to {model_filename}")
```

```python
config = {
    "features": features,
    "classes": class_names
}
with open("model_config.json", "w") as f:
    json.dump(config, f)

print(f"Configuration saved to model_config.json")
print(f"Classes: {class_names}")
print("--- Training Script Finished ---")
```
